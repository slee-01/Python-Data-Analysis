The dataset is available at: https://www.kaggle.com/jboysen/london-crime/download

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('london_crime_by_lsoa.csv')
```

```python
df.sample(5)
```

```python
df.shape
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,6))
df.groupby('borough').size().plot.bar()
plt.title('Number of Crimes by Borough')
```

```python
df.groupby('major_category').size().plot.bar()
plt.title('Number of Crimes by Category')
```

```python
df.groupby('major_category').size().plot.pie()
plt.title('Number of Crimes by Category')
```

```python
df.groupby('year').value.sum().plot.bar()
plt.ylim(650000, 770000)
plt.title('Number of Crimes per Year')
```
