## Data Analysis

### Sales Data Grouped by Year

```python
year = sales.groupby('Year').sum()
year
```

### Sales Trend Comparison by Continent

```python
year.iloc[:, 1:].plot()
```

### Global Sales Trend by Year

```python
gsales = sales.groupby('Year')['Global_Sales'].sum()
gsales.plot(kind='bar')
```

We see that the sales increased until it peaked in 2008, then sharply dropped afterwards. We may want to investigate other sources to find out what factor contributed to the decline in sales after 2008, or check whether this dataset is complete and up to date.

### Most Popular Genre Overall

```python
sales['Genre'].value_counts().sort_values(ascending=False).plot(kind='bar')
```

## Specific Questions

### What year had the most sales in EU?

```python
eu = sales.groupby('Year')['EU_Sales'].sum()
eu.sort_values(ascending=False)
```

It was in 2009, one year after the global peak in 2008.

### What was the most popular platform in EU that year?

```python
sales2008 = sales.loc[sales['Year']==2008]
sales2008.groupby('Platform').sum().sort_values(by='EU_Sales', ascending=False)
```

The most popular platform in EU in 2008 was Wii.

### In what year did EU sales begin to surpass sales in Japan (after 1990)? 

```python
sales.groupby('Year')['NA_Sales'].sum().plot.line()
sales.groupby('Year')['EU_Sales'].sum().plot.line()
sales.groupby('Year')['JP_Sales'].sum().plot.line()
sales.groupby('Year')['Other_Sales'].sum().plot.line()
plt.legend()
plt.show()
```

We re-plot the sales trend comparison by continent from earlier

```python
year.iloc[:, 1:].plot()
```

By visual inspection, we can narrow it down to the range between 1995 and 2000.

```python
df = year['JP_Sales'] - year['EU_Sales']
df.plot(kind='bar')
```

```python
df
```

We see that 1998 was the first year after 1990 that sales in EU began to surpass sales in Japan.
